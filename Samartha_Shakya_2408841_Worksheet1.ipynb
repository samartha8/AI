Problem - 3: Array Operations:
For the following arrays:
x = np.array([[1,2],[3,5]]) and y = np.array([[5,6],[7,8]]);
v = np.array([9,10]) and w = np.array([11,12]);
Complete all the task using numpy:
1. Add the two array.
2. Subtract the two array.
3. Multiply the array with any integers of your choice.
4. Find the square of each element of the array.
5. Find the dot product between: v(and)w ; x(and)v ; x(and)y.
6. Concatenate x(and)y along row and Concatenate v(and)w along column.
{Hint:try np.concatenate() or np.vstack() functions.
7. Concatenate x(and)v; if you get an error, observe and explain why did you get the error?


```python
import numpy as np

# Define the arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays using np.add()
addition = np.add(x, y)
print("Addition of x and y:\n", addition)

# 2. Subtract the two arrays using np.subtract()
subtraction = np.subtract(x, y)
print("\nSubtraction of x and y:\n", subtraction)

# 3. Multiply the arrays by integers using np.multiply()
multiplication_x = np.multiply(x, 3)  # Multiply x by 3
multiplication_y = np.multiply(y, 2)  # Multiply y by 2
print("\nMultiplication of x by 3:\n", multiplication_x)
print("Multiplication of y by 2:\n", multiplication_y)

# 4. Find the square of each element using np.square()
square_x = np.square(x)
square_y = np.square(y)
print("\nSquare of elements in x:\n", square_x)
print("Square of elements in y:\n", square_y)

# 5. Compute dot products using np.dot()
dot_vw = np.dot(v, w)  # Dot product of v and w
dot_xv = np.dot(x, v)  # Dot product of x and v
dot_xy = np.dot(x, y)  # Dot product of x and y
print("\nDot product of v and w:", dot_vw)
print("Dot product of x and v:\n", dot_xv)
print("Dot product of x and y:\n", dot_xy)

# 6. Concatenate x and y along rows using np.concatenate() and v and w along columns using np.vstack()
concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_column = np.vstack((v, w)).T  # Transpose for column-wise concatenation
print("\nConcatenation of x and y along rows:\n", concat_xy_row)
print("Concatenation of v and w along columns:\n", concat_vw_column)

# 7. Attempt to concatenate x and v using np.concatenate() (observe the error)
try:
    concat_xv = np.concatenate((x, v), axis=0)
    print("\nConcatenation of x and v:\n", concat_xv)
except ValueError as e:
    print("\nError while concatenating x and v:", e)
```

Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)

T = BTAT
.

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}


```python
import numpy as np

# Matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

### 1. Prove A.A⁻¹ = I (Identity Matrix)
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("1. A * A⁻¹:\n", identity)  # Should result in Identity matrix

### 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("BA:\n", BA)
print("AB == BA:", np.array_equal(AB, BA))

### 3. Prove (AB)ᵀ = BᵀAᵀ
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)  # BᵀAᵀ
print("\n3. (AB)ᵀ:\n", AB_transpose)
print("BᵀAᵀ:\n", BT_AT)
print("(AB)ᵀ == BᵀAᵀ:", np.array_equal(AB_transpose, BT_AT))

# Represent the system in Matrix form: AX = B
A_eq = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])
B_eq = np.array([-1, -3, 9])

# Solve using np.linalg.inv
A_eq_inv = np.linalg.inv(A_eq)  # Find the inverse of A
X = np.dot(A_eq_inv, B_eq)  # Solve for X using X = A⁻¹B
print("\nSolution using Inverse Method:\n", X)

# Alternatively, solve using np.linalg.solve (numerically more stable)
X_solve = np.linalg.solve(A_eq, B_eq)
print("\nSolution using np.linalg.solve:\n", X_solve)
```
