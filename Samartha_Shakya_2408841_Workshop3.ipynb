```python
import pandas as pd
data = {'Category': ['A','B','A','B','A'],'Value':[10,20,30,40,50]}
df = pd.DataFrame(data)
grouped = df.groupby('Category')['Value'].mean()
print(grouped)
```

```python
import pandas as pd
data = {'Category': ['A','B','A','B','A'],'Value':[10,20,30,40,50]}
df = pd.DataFrame(data)
df['Normalized'] = df.groupby('Category')['Value'].transform(lambda x: x/ x.sum())
print(df)
```

```python
import pandas as pd
data = {'Category': ['A','B','A','B','A'],'Value':[10,20,30,40,50]}
df = pd.DataFrame(data)
filtered = df.groupby('Category').filter(lambda x: x['Value'].sum()>60)
print(filtered)
```

```python
import matplotlib.pyplot as plt

# Data for plotting
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]
steps_last_week = [9788, 8710, 5308, 17630, 21309, 4002, 5223]

# Plotting step counts for this week
plt.plot(days, steps_walked, "o-g", label="This week")

# Plotting step counts for last week
plt.plot(days, steps_last_week, "v--m", label="Last week")

# Adding title and axis labels
plt.title("Step Count | This Week and Last Week")
plt.xlabel("Days of the Week")
plt.ylabel("Steps Walked")

# Adding grid and legend
plt.grid(True)
plt.legend()

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)
t1= np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0,5.0,0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show
```
